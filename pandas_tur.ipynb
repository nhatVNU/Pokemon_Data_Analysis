**FILLTER DATAS**

```python
#Fillter the Pokemon that Fire and HP=50 and Attack= 52
Pokemon=Data_csv.loc[(Data_csv["Type 1"]=='Fire')&(Data_csv["HP"]==50)&(Data_csv["Attack"]==52)]
print(Pokemon.Name)
```

Reset index from filltered

```python
Pokemon_Reset_Index = Pokemon.reset_index()
print(Pokemon_Reset_Index.Name)
```

```python
#Filter the Pokemon that Fire and its name include 'Mega'
Pokemon_Mega = Data_csv.loc[(Data_csv["Type 1"]=="Fire") & (Data_csv.Name.str.contains("Mega"))]
Pokemon_Mega = Pokemon_Mega.reset_index()
print(Pokemon_Mega)
```

```python
#Filter the Pokemon that Fire and its name didn't include 'Mega'
Pokemon_unMega = Data_csv.loc[(Data_csv["Type 1"]=="Fire") & (~Data_csv.Name.str.contains("Mega"))]
Pokemon_unMega = Pokemon_unMega.reset_index()
print(Pokemon_unMega.head(10))
```

```python
#Find Pokemon so that Type 1 = Fire or Grass 
Pokemon_ = Data_csv.loc[(Data_csv["Type 1"].str.contains("Fire|Grass"))]
Pokemon_ = Pokemon_.reset_index()
print(Pokemon_.head(10))
```

```python
#Tim pokemon ma ten bat dau bang chu "Pi"
import re #regular expression dung de search unicode string va 8 bit strings
_Pokemon = Data_csv.loc[(Data_csv["Name"].str.contains("^Pi[a-z]*", regex=True))]
# '^' để kí hiệu từ bắt đầu bằng Pi
# '[a-z]' các chữ cái tiếp theo bắt đầu từ a đến z
# '*' Dùng để tìm kiếm hết tất cả thỏa điều kiện
print(_Pokemon.head(20))
```

**Thay đổi dữ liệu của một dữ kiện nào đó**

```python
# Những pokemon có attack > 50 và hệ cỏ thì speed và defense đều =0
Data_csv.loc[(Data_csv["Attack"]>50) & (Data_csv["Type 1"]=="Grass"), ["Speed", "Defense"]]= 0
Data_csv.head(20)
```

**MATH**

- Average (avg):

```python
import pandas as pd 
data=pd.read_csv("D:\Python File\Pandas_Tutorial\Sample_data\pokemon_data.csv")
mean_data = data.groupby(data["Type 1"]).mean().sort_values("Attack", ascending=False)
mean_data 
 
```

- (sum): 

```python
sum_data = data.groupby(data["Type 1"]).sum().sort_values("Attack", ascending=False)
sum_data 
```

- (count) 


```python
count_data = data.groupby([data["Type 1"], data["Type 2"]]).count()
count_data.head(10)
```
